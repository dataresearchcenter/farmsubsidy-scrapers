```python
import pandas as pd
```

```python
YEAR = 2017
```

```python
def get_year(year):
    df = pd.read_csv('data/dk_%s_raw.csv.gz' % year, encoding='latin-1', delimiter=';', compression='gzip')
    df['year'] = year
    return df

df = get_year(YEAR)
df.head()
```

```python
df['Navn'].isnull().sum()
```

```python
df.tail()
```

```python
df = df.dropna(subset=['Navn'])
```

```python
df['year'].value_counts()
```

```python
df[df['Navn'].str.contains('Anonym')]
```

```python
import re
from slugify import slugify

def use_id(row):
    if pd.notnull(row['CVR/SE nr.']):
        return 'DK-%s' % row['CVR/SE nr.']
    return 'DK-%s-%s' % (row['Postnr'], slugify(row['Navn']))

NUM_RE = re.compile('\[Anonym - (\d+)\]')
get_id = lambda x: NUM_RE.match(x).group(1)
df['recipient_id'] = df.apply(lambda x: 'DK-%s' % get_id(x['Navn']) if x['Navn'].startswith('[Anonym - ') else
                              use_id(x), 1)
df['Navn'] = df['Navn'].apply(lambda x: '' if x.startswith('[Anonym - ') else x)
df.head()
```

```python
df['recipient_location'] = df.apply(lambda x: '%s, %s' % (x['Kommune'], x['Region']), 1)

df = df.rename(columns={
    'Navn': 'recipient_name',
    'Postnr': 'recipient_postcode',
    'Measure': 'scheme',
    'EU(DKK)': 'amount'
})

df['currency'] = 'DKK'
df['country'] = 'DK'
df['year'] = YEAR
df.head()
```

```python
df = df.drop(['CVR/SE nr.', 'Kommune', 'Region', 'National(DKK)', 'Sum(DKK)', 'Totalt(DKK)', 'Unnamed: 10'], 1)
df.head()
```

```python
len(df)
```

```python
df['amount'] = pd.to_numeric(df['amount'].str.replace('.', '').str.replace(',', '.'))
```

```python
df['amount'].sum()
```

```python
df.to_csv('dk_%s.csv.gz' % YEAR, index=False, encoding='utf-8', compression='gzip')
```
